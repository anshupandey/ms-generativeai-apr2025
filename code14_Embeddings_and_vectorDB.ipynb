# Exploring Embeddeings from OpenAI

```python
import os
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
v1,v2,v3 = embeddings.embed_documents(["I like python programming, its a good programming language.",
                                    "Python is an amazing programming language, i like it a lot.",
                                    "Hello World, how are you?"])
```

```python
print(v1)
```

```python
len(v1), len(v2), len(v3)
```

```python
from numpy.linalg import norm
import numpy as np
# cosine similarity
np.dot(v1,v2)/(norm(v1),norm(v2))
```

```python
from numpy.linalg import norm
import numpy as np

np.dot(v1,v3)/(norm(v1),norm(v3))
```

```python
from numpy.linalg import norm
import numpy as np

np.dot(v2,v3)/(norm(v2),norm(v3))
```

## Vector Database: **ChromaDB**

```python
import chromadb
chroma_client = chromadb.Client()
```

```python
collection = chroma_client.create_collection(name="my_collection")
```

```python
collection.add(
    embeddings=[[1.2, 2.3, 4.5], [6.7, 8.2, 9.2]],
    documents=["Python is a programming language", "langchain is an framework"],
    metadatas=[{"source": "my_source"}, {"source": "my_source"}],
    ids=["id1", "id2"]
)
```

```python
results = collection.query(
    query_embeddings=[[0.1, 2.1, 3.1]],
    n_results=1
)
results
```
